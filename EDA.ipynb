```python
import pandas as pd
import matplotlib.pyplot as plt
import  seaborn as sns

sns.set(style="whitegrid",palette="muted")

df = pd.read_csv("audiobook_data_2.csv")
df.rename(columns={"Unnamed: 0": "row_id"}, inplace=True)

df.head
```

```python
print("Shape:",df.shape)
print("\nColumn Names:",df.columns.tolist())
print("\nData Types:\n")
print(df.dtypes)
```

```python
print("Missing values:\n",df.isna().sum())
print("\nDuplicate rows:",df.duplicated().sum())
```

```python
df["Target"].value_counts().plot(kind="bar", color=sns.color_palette("muted"))
plt.title("Churn Distribution ( 0= No Churn, 1= Churn )")
plt.xlabel("Target")
plt.ylabel("Number of Customers")
plt.show()
```

```python
df.describe().T
```

```python
df["Minutes_listened"].plot(kind="hist",bins=30,color="skyblue")
plt.title("Distribution of Minutes Listened")
plt.xlabel("Minutes Listened")
plt.ylabel("Frequency")
plt.show()

sns.boxplot(x="Target", y="Completion",data=df)
plt.title("Completion Rate by Churn Status")
plt.show()
```
